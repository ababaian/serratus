### Update Permissions


````bash
# Log-in as lovelywater IAM (EC2)
aws configure
aws configure set default.s3.max_concurrent_requests 100
mkdir tmp; cd tmp

# Install parallel
(wget -O - pi.dk/3 || lynx -source pi.dk/3 || curl pi.dk/3/ || \
  fetch -o - http://pi.dk/3 ) > install.sh
bash install.sh
# rm -rf *

# Download / Parse filelist
aws s3 cp s3://serratus-public/notebook/200613_ab/FilesInSumButNotDb.txt ./

sed 's/\./\t/g' FilesInSumButNotDb.txt \
  | sed 's/\//\t/g' - \
  |  cut -f2 - \
  > sra.update


# Update Summaries
nSRA=$(wc -l sra.update | cut -f1 -d' ')

yes aws s3api put-object-acl --acl "public-read" --bucket lovelywater \
  --key summary/ | head -n $nSRA > summary.1
  
yes .summary | head -n $nSRA \
  | paste summary.1 sra.update - \
  | tr -d '\t' \
  > sra.summary


# Update Bam files
yes aws s3api put-object-acl --acl "public-read" --bucket lovelywater \
  --key bam/ | head -n $nSRA > bam.1
  
yes .bam | head -n $nSRA \
  | paste bam.1 sra.update - \
  | tr -d '\t' \
  > sra.bam

rm bam.1 summary.1
````
