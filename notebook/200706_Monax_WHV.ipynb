## Introduction

While coronaviruses are well characterized, other viral families offer higher probability for identifying novel strains. In a personal correspondance with D. Streiker following a [comment on the Serratus virological post](https://virological.org/t/serratus-the-ultra-deep-search-to-discover-novel-coronaviruses/516) and in extension to the recent pre-print from [Bergner et al.](https://www.biorxiv.org/content/10.1101/2020.06.17.156745v2) we were interested if we could identify novel deltaviruses.

A recent develpoment to `Serratus` is the use of the `blastx` algorithm for a search against a protein reference as opposed to nucleotide `bowtie2`. This in theory, and in practice allows for the detection of deeper homologies of distant strain/species. The pilot search included viral UniProt sequences and all known deltavirus antigen proteins, as well as two representative dAg from vampire bats provided by Streiker and collegues, and dAg from the bambi sample. 100K sequencing libraries from non-human mammals were selected as the pilot search space [mamm_SraRunInfo.csv](SRA-queries).

With this we were able to identify two novel deltaviruses; Murray (`SRR2136906`) in _Marmota monax_ (Eastern woodchuck) and Lassie (`SRR7910143`) in _Peropteryx macrotis_ (Lesser dog-like bat).

## Murray Virus / Woodchuck Deltavirus (WDV)

The murray virus was assembled from `SRR2136906` to yield [a 1753 nt](200706_ab/murray.fa) circular genome. The virus has the characteristic 'bar' folding of a deltavirus.

![Figure 2: Mfold base-pairing of Murray Virus](200706_ab/murray_mfold.png)

The groundhog samples from this study by [Fletcher et al., 2015](https://journals.plos.org/plospathogens/article?id=10.1371/journal.ppat.1005103) on the effects of IFNa on experimental infection with Woodchuck Hepatitis Virus (WHV). `SRR2136906` and `SRR2136907` samples with high titre Murray are replicate liver samples from the placebo-treated (WHV, no IFNa) animal designated `1008` taken on week 1 of the study. The same animal `1008` shows trace amounts of Murray virus at weeks 4, 8, 12, and 16 and no deltavirus reads on weeks 3, 20 and 24 suggesting waning viral levels following WHV challenge.  WHV infection is confirmed in the [summary](https://lovelywater2.s3.amazonaws.com/summary/SRR2136906.summary) file for `1008`. Within this study the placebo treated animal `1016` shows trace Murray virus on week 24.

Similiar to how human deltavirus (HDV) is a satellite virus that requires hepetitis virus B (HBV) for it's replication, it is likely that Murray is a satellite virus requiring WHV for it's replication, although this will have to be tested experimentally to be validated. The woodchuck is a model organism for hepatocellular carcionoma (HCC) caused by WHV, the inconsistent precence of Murray virus and it's possible effects on HCC development in this model system may be a confounding variable complicating studies using this model system.

````bash
#!/bin/bash
#### CONTAINER WORKFLOW
yum install -y vim less gzip wget unzip tar

mkdir align; cd align

# Install HISAT2
wget ftp://ftp.ccb.jhu.edu/pub/infphilo/hisat2/downloads/hisat2-2.1.0-Linux_x86_64.zip
unzip hisat2-2.1.0-Linux_x86_64.zip
mv hisat2-2.1.0/* /usr/bin/ 

# Install samtools
SAMTOOLSVERSION='1.10'
wget -O /samtools-${SAMTOOLSVERSION}.tar.bz2 \
  https://github.com/samtools/samtools/releases/download/${SAMTOOLSVERSION}/samtools-${SAMTOOLSVERSION}.tar.bz2 &&\
  tar xvjf /samtools-${SAMTOOLSVERSION}.tar.bz2 && rm /samtools-${SAMTOOLSVERSION}.tar.bz2

cd samtools-${SAMTOOLSVERSION} && make && make install

cd .. && rm -rf samtools-${SAMTOOLSVERSION}

#download monax genome
wget https://ftp.ncbi.nlm.nih.gov/genomes/all/GCA/901/343/595/GCA_901343595.1_MONAX5/GCA_901343595.1_MONAX5_genomic.fna.gz
#should work
gzip -d GCA_901343595.1_MONAX5_genomic.fna.gz
mv GCA_901343595.1_MONAX5_genomic.fna monax5.fa

# Download MHV genome NC_004107.1
# https://www.ncbi.nlm.nih.gov/nuccore/22256030
# vim mhv.fa

# Download MDV genome (murray)
# https://raw.githubusercontent.com/wiki/ababaian/serratus/assets/murray.fa
# vim mdv.fa

# Make reference genome
cat monax5.fa mhv.fa mdv.fa > monax5v.fa

# Index reference
hisat2-build --seed 1337 -p 36 monax5v.fa monax5v
````

````bash
#!/bin/bash
#read the sras file into array
# vim sra.txt
mapfile -t SRARRAY < sra.txt

# Upload to S3
S3_PATH=s3://serratus-public/notebook/200702_ab
aws s3 sync ./ "$S3_PATH"/

# Align each SRA
for SRA in "${SRARRAY[@]}"
do
   mkdir -p $SRA
   cd $SRA

   RGLB=$SRA
   echo "Aligning $SRA"

   #hisat2 command - random things for rg-id, etc
   hisat2 --rg-id "$SRA" --rg LB:"$SRA" \
          --rg SM:"woodchuck" --rg PL:"ILLUMINA" \
          --rg PU:"xxx" -p 35 -x ../monax5v \
          --sra-acc $SRA | samtools view -bS - \
          > aligned_unsorted.bam

   #sort bam file
   samtools sort aligned_unsorted.bam > "$SRA".monax5v.bam
   samtools index "$SRA".monax5v.bam
   rm aligned_unsorted.bam

   samtools flagstat "$SRA".monax5v.bam > "$SRA".flagstat

   # Upload to S3
   aws s3 sync ./ "$S3_PATH"/"$SRA"/
   cd ..
done
````

````bash
# Correct error in first version
# re-do flagstats
for SRA in $(ls | grep "SRR"); do
  cd $SRA
  samtools flagstat -@ 35 "$SRA".monax5v.bam > "$SRA".flagstat
  cd ..
done 
````
